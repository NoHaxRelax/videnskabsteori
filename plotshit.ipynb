```python
data = []

with open('datatinng.txt', 'r') as file:
    for line in file:
        data.append(line.split(' ')[1])

print(data)
```

```python
import matplotlib.pyplot as plt

changeddata = []
for i in range(-1, 2):
    changeddata.append(data.count(str(i)))

print(changeddata)

plt.bar(range(-1,2), changeddata, color=["maroon", "gold", "seagreen"])
plt.ylabel('Number of statements')
plt.xticks(range(-1,2), ['Negative', 'Neutral', 'Positive'])

plt.show()
```
