```python
data = []

with open('datatinng2.txt', 'r') as file:
    for line in file:
        data.append(line.split(' ')[1])

print(data)
```

```python
import matplotlib.pyplot as plt

changeddata = []
for i in range(-1, 2):
    changeddata.append(data.count(str(i)))

print(changeddata)

plt.bar(range(-1,2), changeddata, color=["maroon", "gold", "seagreen"])
plt.ylabel('Number of statements')
plt.xticks(range(-1,2), ['Negative', 'Neutral', 'Positive'])
plt.legend()

plt.show()
```

```python
# Calculate total and percentages
total = sum(changeddata)
percentages = [count/total*100 for count in changeddata]

# Create the plot
plt.figure(figsize=(8, 6))
bars = plt.bar(range(-1, 2), changeddata, color=["maroon", "gold", "seagreen"], edgecolor='black', linewidth=1.5)
plt.ylabel('Number of statements', fontsize=12, fontweight='bold')
plt.title('Sentiment Analysis Results', fontsize=14, fontweight='bold', pad=20)
plt.xticks(range(-1, 2), ['Negative', 'Neutral', 'Positive'], fontsize=11)
plt.yticks(fontsize=10)
plt.ylim((0,200))

# Add grid for better readability
plt.grid(axis='y', alpha=0.3, linestyle='--')

# Annotate each bar with count and percentage
for i, (bar, count, percentage) in enumerate(zip(bars, changeddata, percentages)):
    height = bar.get_height()
    # Position text slightly above the bar
    plt.text(bar.get_x() + bar.get_width()/2, height + max(changeddata)*0.01,
             f'{count}\n({percentage:.1f}%)',
             ha='center', va='bottom', fontsize=11, fontweight='bold')

# Add total count annotation
plt.text(0.5, 0.95, f'Total Statements: {total}',
         transform=plt.gca().transAxes,
         ha='center', fontsize=12, fontweight='bold',
         bbox=dict(boxstyle='round,pad=0.5', facecolor='lightgray', alpha=0.7))

plt.tight_layout()
plt.show()
```
